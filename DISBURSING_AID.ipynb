```python
from pandas.io.parsers.readers import read_csv
df=read_csv('Loan.csv')
```

```python
df.head(30)
```

```python
# data cleaning and missing value
df.isnull()
# the output must be in the form of true and false
# true is coming where ever there is a missing value for eg NaN value
```

```python
df.shape
```

```python
total = df.isnull().sum()
```

```python
total
```

```python
df['Gender'].value_counts()
```

```python
df.Gender = df.Gender.fillna('Male')
```

```python
df['Married'].value_counts()
```

```python
df.head(30)
```

```python
df.Married = df.Married.fillna('yes')
```

```python
df['Dependents'].value_counts()
```

```python
df.Dependents = df.Dependents.fillna('0')
```

```python
df['Self_Employed'].value_counts()
```

```python
df.Self_Employed = df.Self_Employed.fillna('No')
```

```python
mean = df.LoanAmount.mean()
mean
```

```python
df.LoanAmount = df.LoanAmount.fillna(mean)
```

```python
df['Loan_Amount_Term'].value_counts()
```

```python
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(360.0)
```

```python
df['Credit_History'].value_counts()
```

```python
df.Credit_History = df.Credit_History.fillna(1.0)
```

```python
total = df.isnull().sum()
```

```python
total
```

```python
df.head(30)
```

```python
# converting categorical variable
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
for i in var_mod:
  df[i] = le.fit_transform(df[i])
```

```python
df.head(10)
# df.tail()
```

```python
# feature selection area
# dividing into independent and dependent variable.
x = df.iloc[:,1:12]
y = df.iloc[:,12]
```

```python
x.head()
```

```python
y.head()
```

```python
# splitting data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.40,random_state=0)
```

```python
X_test
```

```python
# building decision tree model :
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
y_predictions = model.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test,y_predictions)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_predictions))
```

```python
# import libraries for SVM
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train the SVM model
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train, y_train)
```

```python
# Make predictions using SVM
svm_predictions = svm_model.predict(X_test)
```

```python
y_predictions
```

```python
y_test
```

```python
# Evaluate SVM model
svm_accuracy = accuracy_score(y_test, svm_predictions)
print("SVM Accuracy:", svm_accuracy)
print('SVM Classification Report:')
print(classification_report(y_test, svm_predictions))
```

```python
# building logistic regression model
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train,y_train)
```

```python
y_pred=logistic_regression.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test,y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
#Xgboost Model
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Initialize the XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

# Train the model on the training data
xgb_model.fit(X_train, y_train)
```

```python
# Make predictions on the test data
xgb_predictions = xgb_model.predict(X_test)
```

```python
# Calculate accuracy
xgb_accuracy = accuracy_score(y_test, xgb_predictions)
print("XGBoost Accuracy:", xgb_accuracy)

# Print classification report
print('XGBoost Classification Report:')
print(classification_report(y_test, xgb_predictions))
```

```python
# KNN Model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Initialize the KNN model (you can adjust 'n_neighbors' as needed)
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train the model on the training data
knn_model.fit(X_train, y_train)
```

```python
# Make predictions on the test data
knn_predictions = knn_model.predict(X_test)
```

```python
# Calculate accuracy
knn_accuracy = accuracy_score(y_test, knn_predictions)
print("KNN Accuracy:", knn_accuracy)

# Print classification report
print('KNN Classification Report:')
print(classification_report(y_test, knn_predictions))
```
